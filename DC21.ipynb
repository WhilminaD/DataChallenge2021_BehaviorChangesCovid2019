```python
from scipy.stats import chi2_contingency
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x,y)
    
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
```

```python
cols = ['Sex','Age','Married','PhyAc_After','Exe_After','Groc_Shp','ONL_SHP_AFTR','Outside_Ac','commute_after']
newdata = dataset[cols] 

newdata.head()
```

```python
#check nulls
newdata.isnull().sum()
```

```python
rows= []

for var1 in newdata:
    
    col = []
    for var2 in newdata :
        cramers =cramers_v(newdata[var1], newdata[var2]) # Cramer's V test
        col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
    rows.append(col)


cramers_results = np.array(rows)
df = pd.DataFrame(cramers_results, columns = newdata.columns)



df
```

# Hypothesis Testing

Physical Activity

```python
#H0 - There is no realationship between Marital status and Physical Activity
#H1 - There is realationship between Marital status and Physical Activity
```

```python

#Testing the relationship
crossdata = pd.crosstab(dataset['Married'], dataset['PhyAc_After'])
chi_res = chi2_contingency(crossdata)
print('Chi2 Statistic: {}, p-value: {}'.format(chi_res[0], chi_res[1]))

print(crossdata)

if(chi_res[1]<=0.05):
    print("Reject the null hypothesis")
    print("There is realationship between Marital status and Physical Activity")
else:
    print("Fail to reject the null hypothesis")
    print("There is no realationship between Marital status and Physical Activity")
```

```python
#H0 - There is no realationship between Sex status and Physical Activity
#H1 - There is realationship between Sex and Physical Activity
crossdata = pd.crosstab(dataset['Sex'], dataset['PhyAc_After'])
chi_res = chi2_contingency(crossdata)
print('Chi2 Statistic: {}, p-value: {}'.format(chi_res[0], chi_res[1]))

print(crossdata)

if(chi_res[1]<=0.05):
    print("Reject the null hypothesis")
    print("There is realationship between Sex and Physical Activity")
else:
    print("Fail to reject the null hypothesis")
    print("There is no realationship between Sex status and Physical Activity")
```

```python

#H0 - There is no realationship between Age status and Physical Activity
#H1 - There is realationship between Age and Physical Activity

crossdata = pd.crosstab(dataset['AgeRange'], dataset['PhyAc_After'])
chi_res = chi2_contingency(crossdata)
print('Chi2 Statistic: {}, p-value: {}'.format(chi_res[0], chi_res[1]))

print(crossdata)

if(chi_res[1]<=0.05):
    print("Reject the null hypothesis")
    print("There is relationship between Age status and Physical Activity")
else:
    print("Fail to reject the null hypothesis")
    print("There is no relationship between Age  and Physical Activity")
```
